### Histograms are the most common means of looking at a single variable (Univariate). The values are “binned,” meaning they are grouped together and plotted to show the distribution of the variable.

```python
fig = plt.figure(figsize=(12,12))
axes1 = fig.add_subplot(1,1,1)
axes1.hist(tips['total_bill'], bins=8)
axes1.set_title("Histogram of total bill")
axes1.set_xlabel('total bill')
axes1.set_ylabel('frequency')
```

## Bivariate

### Scatter Plot: used when a continuous variable is plotted against another continuous variable.

```python
scatter_plot = plt.figure(figsize= (10, 10))
axes1 = scatter_plot.add_subplot(1,1,1)
axes1.scatter(tips['total_bill'], tips['tip'])
axes1.set_title('scatter plot of total bill vs tip')
axes1.set_xlabel('total bill')
axes1.set_ylabel('tip')
```

### Boxplot: Used when a discrete is plotted against continuous.

```python
boxplot = plt.figure(figsize=(8,8))
axes1 = boxplot.add_subplot(1,1,1)
axes1.boxplot([tips[tips['sex']== 'Female']['tip'], tips[tips['sex']== 'Male']['tip']], labels= ['Female', 'Male'])
axes1.set_title('Boxplot of Tip by Sex')
axes1.set_xlabel('Sex')
axes1.set_ylabel('Tip')
```

```python
x = tips[tips['sex']== 'Female']['tip']
x.head()
```

```python
def recode_sex(sex):
    if sex == 'Female':
        return 0
    else:
        return 1

tips['sex_color'] = tips['sex'].apply(recode_sex)

scatter_plot = plt.figure(figsize=(8,8))
axes = scatter_plot.add_subplot(1,1,1)
axes.scatter(x= tips['total_bill'], y= tips['tip'], s= tips['size']*10, c= tips['sex_color'], alpha= 0.5)
axes.set_title("total bill vs tip colored by sex and sized by Size")
axes.set_xlabel("total bill")
axes.set_ylabel("tips")
```

# Scatter Plot

```python
scatter, ax = plt.subplots()
ax = sns.regplot(x= 'total_bill', y= 'tip', data= tips)
ax.set_title("scatter plot for total bill vs tip")
ax.set_xlabel("total bill")
ax.set_ylabel("tip")
plt.show()
```

```python
fig = sns.lmplot(x='total_bill', y='tip', data= tips, hue= 'sex', fit_reg= False)
plt.show()
```

```python
joint = sns.jointplot(x='total_bill', y='tip', data= tips)
joint.set_axis_labels(xlabel= 'total bill', ylabel= 'tip')
joint.fig.suptitle("Total bill vs tips")
plt.show()
```

```python
hexbin = sns.jointplot(x= 'total_bill', y= 'tip', data= tips, kind= 'hex')
hexbin.set_axis_labels(xlabel= 'total bill', ylabel= 'tip')
hexbin.fig.suptitle("hexbin plot of total bill vs tips")
plt.show()
```

```python
bar, ax= plt.subplots()
ax = sns.barplot(x= 'total_bill', y= 'time',data= tips)
ax.set_title("bar plot of average total bill during lunch and dinner")
ax.set_xlabel('Time')
ax.set_ylabel('total bill')
plt.show()
```

```python
box, ax= plt.subplots()
ax = sns.boxplot(x= 'time', y= 'total_bill', data= tips)
ax.set_title("boxplot of total bill by time of day")
ax.set_xlabel('time')
ax.set_ylabel('total bill')
plt.show()
```

```python
violin, ax= plt.subplots()
ax= sns.violinplot(x= 'time', y= 'total_bill', data= tips)
ax.set_label("Violin plot for total bill by time of day")
ax.set_ylabel('total bill')
ax.set_xlabel('time')
plt.show()
```

# III. PLOTTING USING PANDAS OBJECT

## Syntax: Dataframe.plot.Plot_Type() or Series.plot.Plot_Type()

## HISTOGRAMS: Series.plot.hist() or Dataframe.plot.hist()

```python
fig, ax= plt.subplots()
ax = tips['total_bill'].plot.hist()
ax.set_xlabel('Total Bill')
plt.show()
```

```python
fig, ax= plt.subplots()
ax = tips[['total_bill', 'tip']].plot.hist(alpha= 0.5, ax=ax, bins= 20)
plt.rcParams['figure.figsize']= [5,5]
plt.show()
```

## Density Plot:  Dataframe.plot.kde()

```python
fig, ax= plt.subplots()
ax = tips['total_bill'].plot.kde()
ax.set_xlabel('Total Bill')
plt.show()
```

## Scatter Plot: Dataframe.plot.scatter(x= col_name, y= col_name)

```python
fig, ax= plt.subplots()
ax = tips.plot.scatter(x= 'total_bill', y='tip', ax=ax)
ax.set_xlabel('Total Bill')
plt.rcParams['figure.figsize']= [20,10]
plt.show()
```

## Hexbin Plot: Dataframe.plot.hexbin(x= col_name, y= col_name)

```python
fig, ax= plt.subplots()
ax = tips.plot.hexbin(x= 'total_bill', y='tip', ax=ax, gridsize= 20)
ax.set_xlabel('Total Bill')
plt.rcParams['figure.figsize']= [8,8]
plt.show()
```

## Box Plot: Dataframe.plot.box()

```python
fig, ax= plt.subplots()
ax = tips.plot.box(ax=ax)
ax.set_xlabel('Total Bill')
plt.rcParams['figure.figsize']= [20,10]
plt.show()
```
